# 최종 추론 실행\n이 노트북은 `test.csv` 파일의 질문들을 읽어와, `RAGPipeline`을 사용하여 답변을 생성하고, `submission.csv` 파일을 생성하는 최종 추론 과정을 보여줍니다.

## 1. 라이브러리 및 데이터 로드

```python
import sys\nimport os\nimport pandas as pd\nfrom tqdm.auto import tqdm\n\n# 프로젝트 루트를 경로에 추가\nsys.path.append(os.path.abspath('..'))\n\nfrom src.rag_pipeline import RAGPipeline\n\n# 질문 데이터 로드\ntest_df = pd.read_csv('../data/test.csv')
```

## 2. RAG 파이프라인 초기화\nRAGPipeline 클래스를 인스턴스화합니다. 이 과정에서 모든 모델(Embedding, LLM, Reranker)과 FAISS 인덱스가 로드 및 초기화됩니다.\n**주의:** 이 셀은 모델을 로드하므로 실행에 몇 분 정도 소요될 수 있습니다.

```python
rag_pipeline = RAGPipeline()
```

## 3. 추론 실행\n`test.csv`의 각 질문에 대해 추론을 실행하고 결과를 리스트에 저장합니다.

```python
predictions = []\nfor question in tqdm(test_df['Question'], desc="Running Inference"):\n    try:\n        answer = rag_pipeline.infer(question)\n        predictions.append(answer)\n    except Exception as e:\n        print(f"Error processing question: '{question[:50]}...'. Error: {e}")\n        # 에러 발생 시 기본값 '0' 또는 다른 적절한 값으로 대체\n        predictions.append("0")
```

## 4. 제출 파일 생성

```python
submission_df = pd.read_csv('../data/sample_submission.csv')\nsubmission_df['Answer'] = predictions\nsubmission_df.to_csv('../submission.csv', index=False, encoding='utf-8-sig')\n\nprint("✅ submission.csv 파일 생성이 완료되었습니다.")\nprint(submission_df.head())
```
